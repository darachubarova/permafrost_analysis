### 1. Площадка R42 (Якутск-Туймаада): Техногенный и песчаный буфер

Площадка R42 расположена на надпойменной террасе (долина Туймаада) в условиях нарушенного ландшафта, близкого к пригородной зоне.

* **Грунты:** Преимущественно супеси и пески. Песок обладает высокой водопроницаемостью.
* **Физика процесса:** Когда наступает аномально жаркое лето (высокий `TDD`), песок прогревается быстро и глубоко, но дождевая или талая влага в нем не задерживается — она мгновенно дренирует (уходит) вниз или стекает в реку *в тот же самый сезон*.
* **Почему нет отложенного влияния (сдвиг -0.053):** В песчаных грунтах R42 не происходит аккумуляции «скрытой влаги» под зиму. Избыточное тепло или холод уходят «здесь и сейчас». Мерзлота на этой площадке реагирует на климат оперативно, не оставляя опасного инерционного следа на следующий год.

### 2. Площадка R42A / R43 (Шестаковка): Едомный комплекс и ледяной замок

Площадка R43 (часто идет в связке с Шестаковским геокриологическим стационаром) заложена на водоразделе, в условиях естественной лиственничной тайги со сплошным мохово-лишайниковым покровом.

* **Грунты:** Тяжелые суглинки, богатые подземными льдами (едомный комплекс). Суглинок, в отличие от песка, удерживает воду мертвой хваткой.

Вот как работает скрытый механизм, который связывает уменьшение протаивания на R43 со всплеском наводнений через год:

1. **Фаза переувлажнения:** Накануне катастрофических лет в ландшафт Шестаковки поступает избыточная влага. Тяжелые суглинки полностью насыщаются водой.
2. **Парадокс протаивания (Почему `ALT` падает):** Вода обладает гигантской теплоемкостью (чтобы прогреть влажную почву, нужно в разы больше тепла, чем для сухой). Плюс плотный слой мха работает как термос. В результате, летом суглинистый грунт на R43 **прогревается хуже и глубина протаивания (`ALT_R43`) оказывается меньше нормы** (отрицательная Z-оценка).
3. **Осеннее замерзание:** Вся эта избыточная влага уходит под зиму и замерзает, превращаясь в сплошной, монолитный подземный лед (лед-цемент).
4. **Весенний удар (Всплеск ЧС):** Весной, когда начинает таять снег, эта перенасыщенная льдом суглинистая почва водораздела Шестаковки превращается в **абсолютно непроницаемое стекло (криогенный водоупор)**. Талые воды не могут впитаться в землю ни на сантиметр. Вся масса талого снега лавиной скатывается по склонам в малые реки, а оттуда — в русло Лены, формируя катастрофические весенние половодья и выбивая лед с образованием заторов.

### Анализ результата для ВКР (Почему это ценно?)

Этот результат гораздо интереснее для научной работы, чем если бы везде было одинаковое количество ЧС. У вас есть четко выраженные **«спокойные годы»** (по 1 событию) и два явных **критических года** катастрофических масштабов: **2013** и особенно **2018** (когда на Лене и притоках одновременно сорвало сразу много гидропостов и затопило улусы от Табаги до Жиганска).

Разное, даже противоположное влияние площадок **R42** и **R43** на количество ЧС в следующем году — это не ошибка в расчетах. Это красивейший географический и мерзлотный факт, который показывает комиссионной защите ВКР, что вы не просто запустили код, а понимаете реальную физику ландшафтов Якутии.

Давайте посмотрим на цифры сдвигов из вашей таблицы еще раз:

* **`ALT_R42_z` = -0.053** (сдвиг практически нулевой, влияния на межгодовой лаг почти нет).
* **`ALT_R43_z` = -0.317** (заметный отрицательный сдвиг — уменьшение протаивания мерзлоты в году $N$ предвещает всплеск ЧС в году $N+1$).

## 1. Киренск (Верхняя Лена): Феномен «очищения» русла и глубина промерзания

**Главный предиктор:** `max_snow_depth_z` (Сдвиг: **`-1.379`**)

**Второй по значимости:** `FDD_z` (Сдвиг: **`-1.021`**)

* **Физическая интерпретация:** Обратите внимание на резкий отрицательный сдвиг высоты снежного покрова в году $N$ (`-1.379`). Это означает, что **всплеску наводнений в году $N+1$ предшествует малоснежная зима в году $N$**.
* **Гидрологический механизм:** В верховьях реки Лены малоснежный год $N$ приводит к тому, что почва и пойменные территории лишаются теплоизоляционного снежного покрова. В сочетании с отрицательным сдвигом `FDD_z` (`-1.021`, что означает высокую суровость зимы) происходит **экстремально глубокое промерзание речного грунта и самого ледяного панциря**. Переходя в год $N+1$, этот монолитный, глубоко промерзший лед не успевает деградировать под первыми лучами солнца и весной образует катастрофические, непреодолимые заторы для паводковой волны.

## 2. Олёкминск (Среднее течение): Зимний фактор как главный предиктор

**Главный предиктор:** `FDD_z` (Сдвиг: **`-1.468`**)

## 6. Почему [`N_t_z`](Dara/temp-solid.txt:35) почти везде близок к нулю

Это один из самых методически важных выводов.

[`N_t`](Dara/temp-solid.txt:35) — отношение летнего тепла, накопленного грунтом, к летнему теплу воздуха. Если сдвиг [`N_t_z`](Dara/temp-solid.txt:35) около нуля, значит:

1. **Летняя связь «воздух — грунт» относительно стабильна.**  
   В тёплый сезон грунт получает тепло от атмосферы примерно пропорционально во всех типах лет.

2. **ЧС с лагом формируются не через изменение летней проводимости тепла.**  
   То есть не видно, что перед всплесками ЧС грунт начинал пропускать летнее тепло существенно эффективнее или слабее.

3. **Гораздо важнее зимний и межсезонный режим.**  
   Сильнее реагируют [`N_f`](Dara/temp-solid.txt:35), [`SO`](Dara/temp-solid.txt:24), [`TO`](Dara/temp-solid.txt:24), [`MAGST`](Dara/temp-solid.txt:14), [`TTOP`](Dara/temp-solid.txt:19). Это означает, что память системы формируется зимой, при промерзании переувлажнённого грунта, а не только летом.

4. **Для ВКР это защищает модель от упрощения.**  
   Можно сказать, что анализ не свёлся к банальному «жаркое лето вызывает ЧС». Если бы так было, [`N_t_z`](Dara/temp-solid.txt:35) тоже должен был бы резко отличаться. Но он почти не отличается. Значит, риск задаётся комбинацией летнего теплового запаса, влаги, снежного покрова и зимнего замерзания.

## 7. Готовая итоговая формулировка для диссертации

Лены определяется не столько изменением эффективности летней передачи тепла в грунт, сколько зимне-межсезонной перестройкой мерзлотного режима. Для Киренска и Олёкминска характерны отрицательные сдвиги [`MAGST_z`](Dara/temp-solid.txt:14) и положительные сдвиги [`TO_z`](Dara/temp-solid.txt:24), отражающие изменение вертикального температурного профиля мерзлоты. >
> Таким образом, близкий к нулю [`N_t_z`](Dara/temp-solid.txt:35) является важным отрицательным результатом: он показывает, что долгосрочные предпосылки ЧС формируются преимущественно не в фазе летнего прогрева, а в фазе последующего осенне-зимнего закрепления влаги и перестройки теплового режима мерзлоты.

## 1. Жиганск: к чему приводит общий тепловой фон?

Для Жиганска положительные сдвиги `MAGST_z = +0.400` и `TTOP_z = +0.530` означают, что перед годами с всплеском ЧС мерзлотный профиль находится в более тёплом состоянии: прогрета и поверхность грунта, и более глубокая зона у кровли мерзлоты.

Это приводит к нескольким последствиям:

1. **Ослабление устойчивости мерзлотного основания.**  
   Если мерзлота теплее, она ближе к фазовому переходу лёд → вода. Даже небольшие оттепели, весенние осадки или резкое потепление могут быстрее нарушить устойчивость грунта и льда.

2. **Рост талого и поверхностного стока.**  
   Тёплая мерзлота хуже работает как стабильный холодный массив. Весной верхний слой быстрее переходит в активное состояние, но глубже сохраняется мерзлотный водоупор. Поэтому вода не уходит глубоко, а стекает по поверхности к руслу.

3. **Повышение ледовитости и нестабильности русловой зоны.**  
   В Жиганске важны также `thaw_d_z = +2.000` и `precip_spring_z = +1.426`. То есть общий тёплый фон сочетается с оттепелями и весенним увлажнением. Это может приводить к образованию наледей, ледяных корок, переувлажнённых промерзающих слоёв и нестабильному вскрытию русла.

4. **Наложение местного стока на транзитную волну с юга.**  
   Нижняя Лена получает паводковую волну сверху. Если в Жиганске местные мерзлотно-гидрологические условия уже нестабильны, то транзитная волна усиливается местным стоком и заторными процессами.

Итог:

> В Жиганске общий положительный тепловой фон мерзлоты приводит к снижению устойчивости мерзлотно-ледового комплекса, усилению поверхностного стока, нестабильному весеннему вскрытию русла и повышению вероятности заторов/подтоплений при приходе паводковой волны сверху по Лене.

## 2. Олёкминск: к чему приводит изменение вертикального температурного контраста и зимней изоляции грунта?

Это приводит к следующему:

1. **Усиление промерзания верхнего слоя и пойменных участков.**

2. **Формирование плотного мерзлотного водоупора.**  
   Если верхний слой сильно промёрз, весной талая вода хуже инфильтрируется. Она начинает стекать по поверхности в русло, усиливая паводочный приток.

4. **Сдвиг сроков естественного очищения русла.**

Итог:

> В Олёкминске изменение вертикального температурного контраста и зимней изоляции грунта приводит к усиленному промерзанию верхнего слоя, формированию прочного ледового покрова, снижению инфильтрации талых вод и задержке естественного очищения русла. В результате весной года N+1 возрастает вероятность мощного заторного всплеска.

## 3. Киренск: к чему приводит то, что более глубокий слой сохраняет аномально высокий тепловой фон?

Это приводит к следующему:

1. **Возникает сильный вертикальный температурный градиент.**  
   Верх промерзает интенсивно, но глубинный слой сохраняет тепло. Это создаёт напряжённое термическое состояние грунта.

2. **Появляются условия для криогенного перераспределения влаги.**  
   При таком градиенте вода может мигрировать к фронту промерзания, замерзать, образовывать ледяные прослои, линзы льда и плотные промёрзшие горизонты.

3. **Формируется многослойный водоупор.**  
   Весной талая вода встречает не просто мерзлоту, а сложную систему промёрзших и ледонасыщенных горизонтов. Инфильтрация резко снижается.

4. **Усиливается поверхностный сток в русло.**  
   Если вода не уходит в грунт, она быстрее поступает в реку. Для верховьев Лены это особенно опасно, потому что именно здесь формируется начальная паводковая волна.

5. **Увеличивается риск динамического вскрытия льда.**  
   В [`Dara/коли-чс в год.txt`](Dara/коли-чс%20в%20год.txt:217) Киренск описан как участок, где малоснежность и промерзание приводят к монолитному льду. Новые параметры показывают, что к этому добавляется глубинная тепловая инерция: лёд и грунт не переходят в весеннее состояние плавно, а сохраняют контрастное, напряжённое состояние.

Итог:

> В Киренске сохранение аномально высокого теплового фона в глубинном слое при охлаждённой поверхности приводит к усилению вертикального температурного градиента, криогенной миграции влаги, образованию ледяных прослоев и водоупорных горизонтов. Весной это усиливает поверхностный сток и способствует резкому, динамическому вскрытию льда с образованием заторов.

## 4. К чему в целом приводят зимний режим промерзания, снежно-почвенная изоляция и вертикальный температурный градиент мерзлоты?

Это общий механизм для всей системы Лены.

Эти факторы приводят не к одному эффекту, а к цепочке:

```text
изменение зимнего промерзания
+ изменение снежной изоляции
+ вертикальный температурный контраст
→ перераспределение влаги в грунте
→ формирование ледяных прослоев и мерзлотного водоупора
→ снижение весенней инфильтрации
→ рост поверхностного стока
→ усиление паводочной волны
→ заторы, подтопления, половодья
```

По участкам это проявляется по-разному:

| Участок | Главный эффект |
|---|---|
| Киренск | глубокое промерзание + вертикальная расслоенность → резкое вскрытие и заторы в верховьях |
| Олёкминск | прочный ледовый покров + промёрзшая пойма → задержка очищения русла и заторный всплеск |
| Якутск | переувлажнение + зимнее цементирование грунта → нулевая инфильтрация и мощный поверхностный сток |
| Жиганск | тёплый мерзлотный фон + оттепели + весеннее увлажнение → нестабильное вскрытие нижней Лены и усиление заторов |

Итоговая формулировка:

> Зимний режим промерзания, снежно-почвенная изоляция и вертикальный температурный градиент мерзлоты определяют, сколько влаги будет законсервировано в деятельном слое в виде льда и насколько проницаемой окажется почва весной. При неблагоприятном сочетании этих факторов формируется криогенный водоупор, который блокирует инфильтрацию талых вод. Это приводит к резкому росту поверхностного стока, ускоренному поступлению воды в русло Лены, усилению паводочной волны и повышению вероятности заторов и подтоплений в следующем году.

Ниже — уточнённый анализ именно в логике «к чему это приводит».

## 1. Жиганск: к чему приводит общий тепловой фон

## 2. Олёкминск: к чему приводит изменение вертикального температурного контраста и зимней изоляции грунта

Для Олёкминска важны отрицательные сдвиги [`MAGST_z`](Dara/temp-solid.txt:14), [`TTOP_z`](Dara/temp-solid.txt:19), положительный сдвиг [`TO_z`](Dara/temp-solid.txt:24) и слабый отрицательный [`N_f_z`](Dara/temp-solid.txt:35). Это означает, что мерзлотный профиль перед ЧС охлаждается, но поверхность охлаждается сильнее, чем глубинный слой.

Это приводит к следующему:

1. Усиливается промерзание верхнего слоя грунта, поймы и прибрежных участков.
2. Формируется более жёсткий мерзлотный водоупор.
3. Весной талая вода хуже впитывается в грунт.
4. Увеличивается поверхностный сток в русло.
5. Формируется более прочный ледовый покров, шуга и подледный наслуд.
6. Русло позже и труднее очищается ото льда.
7. При приходе паводочной волны повышается вероятность заторного всплеска.

Коротко:

> В Олёкминске изменение вертикального температурного контраста и зимней изоляции грунта приводит к усиленному промерзанию пойменно-русловой зоны, формированию прочного льда и снижению инфильтрации. Весной это создаёт условия для задержки очищения русла и мощных заторов.

## 3. Киренск: к чему приводит то, что глубокий слой сохраняет аномально высокий тепловой фон

Для Киренска сочетание отрицательного [`MAGST_z`](Dara/temp-solid.txt:14), положительного [`TTOP_z`](Dara/temp-solid.txt:19), положительного [`SO_z`](Dara/temp-solid.txt:24) и положительного [`TO_z`](Dara/temp-solid.txt:24) особенно показательно. Поверхность холоднее, а глубинный слой теплее. Значит, профиль становится термически расслоенным.

Это приводит к следующему:

1. Возникает сильный вертикальный температурный градиент.
2. Влага начинает перераспределяться по профилю грунта к фронту промерзания.
3. Формируются ледяные линзы, прослои и плотные промёрзшие горизонты.
4. Появляется многослойный мерзлотный водоупор.
5. Весной талая вода почти не инфильтрируется в грунт.
6. Поверхностный сток быстро поступает в верховья Лены.
7. Лёд вскрывается не плавно, а резко и динамически.
8. Возрастает риск формирования заторов уже в верхнем звене речной системы.

Коротко:

> В Киренске сохранение аномально высокого теплового фона на глубине при охлаждённой поверхности приводит к термическому расслоению грунта, миграции влаги, образованию ледяных прослоев и резкому росту поверхностного стока. Это усиливает динамическое вскрытие льда и риск заторов в верховьях Лены.

## 4. К чему в целом приводят зимний режим промерзания, снежно-почвенная изоляция и вертикальный температурный градиент мерзлоты

Эти факторы управляют тем, в каком состоянии грунт и русловая зона войдут в весну следующего года.

Общая цепочка такая:

```text
зимний режим промерзания
+ снежно-почвенная изоляция
+ вертикальный температурный градиент мерзлоты
→ перераспределение влаги в деятельном слое
→ замерзание влаги и образование ледяных прослоев
→ формирование мерзлотного водоупора
→ снижение весенней инфильтрации
→ рост поверхностного стока
→ усиление паводочной волны
→ заторы, подтопления и половодья
```

По станциям это проявляется так:

| Станция | К чему приводит механизм |
|---|---|
| Киренск | к динамическому вскрытию льда и формированию заторов в верховьях |
| Олёкминск | к прочному ледовому покрову, задержке очищения русла и заторному всплеску |
| Якутск | к криогенному цементированию переувлажнённого грунта и нулевой инфильтрации весной |
| Жиганск | к усилению местного стока, нестабильному вскрытию нижней Лены и наложению на транзитную волну |

Итоговая формулировка для ВКР:

> Зимний режим промерзания, снежно-почвенная изоляция и вертикальный температурный градиент мерзлоты определяют способность деятельного слоя принимать талую воду весной. При неблагоприятном сочетании этих факторов в грунте формируется криогенный водоупор. Он блокирует инфильтрацию, усиливает поверхностный сток, ускоряет поступление воды в русло Лены и повышает вероятность заторов, подтоплений и половодий в следующем году.

Для Якутска такой же анализ нужно строить вокруг сочетания: [`TDD_z`](Dara/коли-чс%20в%20год.txt:243), [`precip_z`](Dara/коли-чс%20в%20год.txt:239), [`ALT_R43_z`](Dara/коли-чс%20в%20год.txt:241), а также ваших новых геокриологических параметров [`N_f_z`](Dara/temp-solid.txt:35), [`SO_z`](Dara/temp-solid.txt:24), [`MAGST_z`](Dara/temp-solid.txt:14), [`TO_z`](Dara/temp-solid.txt:24).

## Якутск: к чему приводит сочетание летнего тепла, переувлажнения и зимнего промерзания

Для Якутска в новом выводе примечательны:

| Параметр | Сдвиг | Смысл |
|---|---:|---|
| `TDD_z` | **+0.878** | перед всплесками ЧС лето теплее нормы |
| `N_f_z` | **+0.519** | меняется эффективность зимнего промерзания грунта |
| `TO_z` | **+0.606** | усиливается температурный разрыв между кровлей мерзлоты и поверхностью |

Главная особенность Якутска: здесь механизм не сводится только к «жаркое лето → паводок». Наоборот, получается более сложная цепочка:

```text
тёплое лето
+ летне-осеннее переувлажнение
+ более глубокое протаивание
+ последующее зимнее промерзание влажного слоя
→ криогенное цементирование грунта
→ почти нулевая инфильтрация весной
→ резкий поверхностный сток
→ усиление паводка и риска ЧС
```

## 1. К чему приводит высокий [`TDD_z`](Dara/коли-чс%20в%20год.txt:243) в Якутске

Высокий [`TDD_z`](Dara/коли-чс%20в%20год.txt:243) означает, что в году N тёплый сезон накопил больше тепла. Это приводит к следующему:

1. Увеличивается глубина сезонного протаивания.
2. Деятельный слой становится более мощным и влагоёмким.
3. Верхняя часть мерзлотного профиля может принять больше дождевой и талой воды.
4. При летне-осенних осадках вода проникает глубже в сезонно-талый слой.
5. К зиме формируется большой объём переувлажнённого грунта.

Коротко:

> В Якутске высокий [`TDD_z`](Dara/коли-чс%20в%20год.txt:243) приводит к увеличению мощности деятельного слоя. Такой слой способен накопить больше влаги, которая затем уходит в зиму и превращается в ледяной водоупор.

## 2. К чему приводит отрицательный [`MAGST_z`](Dara/temp-solid.txt:14)

На первый взгляд отрицательный [`MAGST_z`](Dara/temp-solid.txt:14) может казаться противоречием: если лето тёплое, почему поверхность грунта холоднее? Но это как раз важный геокриологический сигнал.

Он может означать, что:

1. Грунт переувлажнён.
2. Влажный грунт имеет большую теплоёмкость.
3. На его прогрев требуется больше энергии.
4. Поэтому средняя температура поверхности грунта может быть ниже, хотя общий тепловой запас сезона высокий.
5. Значительная часть энергии уходит не просто на нагрев, а на фазовые процессы: оттаивание, удержание влаги, последующее замерзание.

Коротко:

> В Якутске отрицательный [`MAGST_z`](Dara/temp-solid.txt:14) приводит к выводу, что перед всплесками ЧС поверхность грунта не просто прогревается, а переходит в переувлажнённое, теплоёмкое состояние. Такой грунт медленнее нагревается, но накапливает влагу, которая зимой цементируется льдом.

## 3. К чему приводит положительный [`N_f_z`](Dara/temp-solid.txt:35)

[`N_f`](Dara/temp-solid.txt:35) показывает, какая доля зимнего холода воздуха реально проходит в грунт. Положительный сдвиг [`N_f_z`](Dara/temp-solid.txt:35) в Якутске означает, что перед всплесками ЧС зимнее промерзание грунта становится более эффективным или более резко выраженным.

Это приводит к следующему:

1. Переувлажнённый деятельный слой промерзает сильнее.
2. Вода в порах грунта превращается в лёд.
3. Образуются ледяные прослои, линзы и цементированные горизонты.
4. Грунт теряет водопроницаемость.
5. Весной талая вода не может уйти вниз.
6. Поверхностный сток резко возрастает.

Коротко:

> В Якутске положительный [`N_f_z`](Dara/temp-solid.txt:35) приводит к усиленному промерзанию переувлажнённого деятельного слоя. Это превращает грунт в криогенный водоупор, который весной блокирует инфильтрацию талой воды.

## 4. К чему приводит положительный [`TO_z`](Dara/temp-solid.txt:24)

[`TO`](Dara/temp-solid.txt:24) — это разница между температурой кровли мерзлоты и температурой поверхности грунта. Положительный [`TO_z`](Dara/temp-solid.txt:24) означает усиление вертикального температурного контраста.

Для Якутска это приводит к следующему:

1. В профиле возникает выраженная термическая неоднородность.
2. Верхний слой и глубинный слой реагируют на климат не одинаково.
3. Влага перераспределяется по профилю к зонам промерзания.
4. Возникают условия для образования ледяных прослоев.
5. Усиливается эффект «ледяного замка».
6. Весной деятельный слой становится плохо проницаемым.

Коротко:

> В Якутске положительный [`TO_z`](Dara/temp-solid.txt:24) приводит к усилению вертикального температурного контраста, миграции влаги и формированию ледяных горизонтов. Это дополнительно укрепляет криогенный водоупор перед весной года N+1.

## 5. К чему приводит положительный [`SO_z`](Dara/temp-solid.txt:24)

[`SO`](Dara/temp-solid.txt:24) отражает разницу между температурой поверхности грунта и температурой воздуха. Его положительный сдвиг в Якутске небольшой, но методически важный.

Он показывает, что перед всплесками ЧС меняется связь «воздух — поверхность грунта». Это может быть связано со снегом, влажностью, растительностью, городским влиянием и теплоизоляцией поверхности.

Это приводит к следующему:

1. Температура воздуха уже не полностью объясняет состояние поверхности грунта.
2. Поверхностный слой начинает вести себя как самостоятельный теплоёмкий буфер.
3. Снег и влажный грунт меняют передачу тепла и холода.
4. Зимнее промерзание становится неоднородным.
5. Возникают условия для локального ледообразования и водоупора.

Коротко:

> В Якутске положительный [`SO_z`](Dara/temp-solid.txt:24) указывает на изменение теплообмена между воздухом и поверхностью грунта. Это приводит к буферизации поверхности, неоднородному промерзанию и усилению условий для криогенного цементирования.

## 6. К чему приводит дефицит снега [`max_snow_depth_z = -0.563`](Dara/permafrost_analysis.ipynb:1923)

Для Якутска отрицательный сдвиг высоты снега означает, что перед всплесками ЧС снежный покров ниже, чем в безопасные годы.

Это приводит к следующему:

1. Снег хуже изолирует грунт от зимнего холода.
2. Холод легче проникает в переувлажнённый деятельный слой.
3. Влага промерзает глубже и плотнее.
4. Формируется более мощный ледоцементированный горизонт.
5. Весной инфильтрация талой воды снижается.
6. Поверхностный сток увеличивается.

Коротко:

> В Якутске дефицит снега приводит к тому, что переувлажнённый после тёплого лета грунт сильнее промерзает зимой. Это усиливает образование мерзлотного водоупора и повышает риск паводочного стока весной.

## 7. Итоговая цепочка для Якутска

Для Якутска механизм выглядит так:

```text
тёплое лето N
→ увеличение мощности деятельного слоя
→ накопление влаги в грунте
→ отрицательный MAGST_z как признак теплоёмкого переувлажнённого слоя
→ дефицит снежной изоляции
→ усиленное зимнее промерзание, положительный N_f_z
→ рост вертикального температурного контраста, положительный TO_z
→ криогенное цементирование грунта
→ блокировка инфильтрации весной N+1
→ резкий поверхностный сток
→ усиление паводка, заторов и подтоплений у Якутска
```

## Готовая формулировка для ВКР

> Для Якутска механизм долгосрочного формирования гидрологических ЧС определяется сочетанием летнего теплового накопления и последующего зимнего закрепления влаги в деятельном слое. Положительный сдвиг [`TDD_z`](Dara/коли-чс%20в%20год.txt:243) указывает на увеличение мощности сезонно-талого слоя в году N, что повышает его способность аккумулировать влагу. Одновременно отрицательный сдвиг [`MAGST_z`](Dara/temp-solid.txt:14) свидетельствует о теплоёмком, переувлажнённом состоянии поверхностного грунта, при котором значительная часть тепловой энергии расходуется на фазовые процессы и удержание влаги. В зимний период положительный сдвиг [`N_f_z`](Dara/temp-solid.txt:35) и дефицит снежного покрова способствуют более эффективному промерзанию этого переувлажнённого слоя. Положительный сдвиг [`TO_z`](Dara/temp-solid.txt:24) отражает усиление вертикального температурного контраста и формирование ледонасыщенных горизонтов. В результате к весне года N+1 формируется криогенный водоупор, который блокирует инфильтрацию талых вод, усиливает поверхностный сток в русло Лены и повышает риск половодий, заторов и подтоплений в районе Якутска.

С учётом добавленных [`ros_event_days`](Dara/permafrost_analysis.ipynb), [`ros_trigger`](Dara/permafrost_analysis.ipynb), [`ros_precip_sum`](Dara/permafrost_analysis.ipynb) и [`ros_precip_max`](Dara/permafrost_analysis.ipynb) главный вывод такой:

> [`Rain-on-Snow`](Dara/permafrost_analysis.ipynb) в синхронном анализе оказался **не универсальным объяснителем крупных региональных всплесков ЧС**, а скорее **локальным оперативным триггером отдельных паводковых лет**.  
> Для катастрофических лет [`2013`](Dara/коли-чс%20в%20год.txt:11) и особенно [`2018`](Dara/коли-чс%20в%20год.txt:13) решающими остаются не ROS-параметры, а сочетание суровой зимы, состояния снежного покрова и весеннего увлажнения.

## 1. Что принципиально изменилось после добавления ROS

Раньше в [`Dara/коли-чс в год.txt`](Dara/коли-чс%20в%20год.txt:159) синхронный анализ интерпретировался через три основных блока:

1. [`FDD_z`](Dara/коли-чс%20в%20год.txt:161) — суровость зимы и прочность ледового покрова.
2. [`precip_spring_z`](Dara/коли-чс%20в%20год.txt:173) — весеннее увлажнение как оперативный триггер.
3. [`max_snow_depth_z`](Dara/коли-чс%20в%20год.txt:187) — запас снега и локальный сток.

Добавление ROS-параметров позволило проверить более тонкую гипотезу:

> возможно, ЧС возникают не просто из-за весенних осадков, а именно из-за дождя по снегу, когда жидкие осадки резко ускоряют таяние снежного покрова и формируют быстрый поверхностный сток.

Но новые таблицы показывают, что эта гипотеза подтверждается **только частично**.

Во всех четырёх станциях средний сдвиг ROS-показателей для категории **«Всплеск ЧС > 1» минус «Безопасно 0 ЧС»** либо отрицательный, либо слабый:

| Станция | [`ros_event_days_z`](Dara/permafrost_analysis.ipynb) | [`ros_precip_sum_z`](Dara/permafrost_analysis.ipynb) | [`ros_precip_max_z`](Dara/permafrost_analysis.ipynb) | Вывод |
|---|---:|---:|---:|---|
| Киренск | -0.953 | -0.538 | -0.108 | ROS не усиливает всплески |
| Олёкминск | -0.233 | -0.168 | -0.147 | ROS слабый и вторичный |
| Якутск | -0.457 | -0.672 | -0.727 | ROS не является причиной пиков |
| Жиганск | -1.148 | -1.260 | -0.853 | ROS в пиковые годы ниже, чем в безопасные |

То есть для крупных системных всплесков [`2013`](Dara/коли-чс%20в%20год.txt:11) и [`2018`](Dara/коли-чс%20в%20год.txt:13) ROS-сигнал не усиливается. Это очень важный отрицательный результат.

## 2. Почему ROS не стал главным фактором катастрофических лет

Ключевая причина — сами пиковые годы.

## 7. Как теперь правильно интерпретировать ROS в работе

ROS стоит оставить в модели, потому что он дал важный научный результат. Но интерпретировать его нужно аккуратно.

### Неправильная формулировка

> Дождь по снегу является главным триггером катастрофических ЧС на Лене.

Эта формулировка не подтверждается таблицами.

### Правильная формулировка

> Дождь по снегу является локальным оперативным триггером отдельных паводковых ситуаций, однако в масштабных региональных всплесках ЧС 2013 и 2018 гг. его роль оказалась вторичной. Катастрофические годы лучше объясняются предшествующим термическим состоянием зимы, прочностью ледового покрова, весенним увлажнением и пространственным наложением паводковой волны по течению Лены.

## 8. Научный смысл отрицательных ROS-сдвигов

Отрицательные сдвиги ROS-параметров — это не ошибка, а полезный результат.

Они показывают, что:

1. ROS-события могут происходить и в безопасные годы.
2. Сам факт дождя по снегу не гарантирует ЧС.
3. Для ЧС важно, в каком состоянии находится система до ROS:
   - насколько прочен лёд;
   - насколько промёрзла пойма;
   - есть ли запас снега;
   - насколько высок весенний приток;
   - совпадает ли местный сток с транзитной волной.
4. Если ледово-мерзлотная система не подготовлена к катастрофе, даже сильный ROS может пройти без ЧС.
5. Если система уже критически подготовлена, ЧС может произойти и без выраженного ROS.

Это можно назвать эффектом **предусловленности ROS**:

> ROS опасен не сам по себе, а только при совпадении с заранее сформированным критическим ледово-мерзлотным состоянием бассейна.

## Почему это НЕ триггер всплеска

Сдвиг считается так:

```text
Сдвиг = среднее значение перед/в годы всплеска ЧС - среднее значение в безопасные годы
```

То есть для `ros_event_days_z`:

```text
-0.915 - 0.038 = -0.953
```

А для `ros_precip_sum_z`:

```text
-0.635 - (-0.097) = -0.538
```

Это значит:

> В годы всплеска ЧС количество ROS-дней и сумма ROS-осадков были ниже, чем в безопасные годы.

Если бы ROS был главным триггером всплеска, мы ожидали бы обратную картину:

```text
Всплеск ЧС > Безопасно
```

То есть сдвиг должен был бы быть **положительным**, например `+0.5`, `+1.0`, `+1.5`.

Но у тебя он отрицательный:

```text
ros_event_days_z = -0.953
ros_precip_sum_z = -0.538
```

Следовательно, ROS не усиливается в годы катастрофических пиков.

## Важный нюанс: ROS может быть связан с обычными ЧС

Обрати внимание на колонку **«Норма 1 ЧС»**:

```text
ros_event_days_z = 0.200
ros_precip_sum_z = 0.231
```

Они положительные.

Это очень важный результат. Он означает:

> ROS может сопровождать годы с единичными ЧС, но не объясняет переход к катастрофическому всплеску >1 ЧС.

То есть правильная интерпретация такая:

* для **локального паводка** ROS может быть триггером;
* для **регионального всплеска ЧС** ROS не является главным фактором;
* катастрофические годы формируются не одним дождём по снегу, а заранее подготовленным состоянием ледово-мерзлотной системы.

## Как это объяснить физически

Rain-on-Snow действительно потенциально опасен: дождь по снегу может ускорить таяние, увеличить поверхностный сток и вызвать локальный паводок. В литературе ROS также рассматривается как фактор быстрого снеготаяния и повышения риска паводков в зимне-весенний период [2].

Но в твоей выборке крупные всплески ЧС, особенно 2013 и 2018 годы, вероятно, были вызваны не ROS, а сочетанием других условий:

```text
суровая зима
→ прочный ледовый покров
→ промёрзшая пойма
→ весенний приток воды
→ задержка вскрытия русла
→ заторы и подтопления
```

То есть ROS — это кратковременный оперативный фактор. А всплеск ЧС — это системное событие, для которого нужна предварительная подготовка бассейна.

## Почему `max_snow_depth_z` положительный именно в Олёкминске

Для Олёкминска синхронная картина такая:

```text
FDD_z = -0.910
precip_spring_z = +1.652
max_snow_depth_z = +0.497
```

Это сочетание читается так:

```text
суровая зима
+ повышенный запас снега
+ сильные весенние осадки
→ резкий рост стока
→ нагрузка на ледовый покров
→ заторы и ЧС
```

То есть положительный снег здесь означает:

1. **Накоплен больший запас воды в снежном покрове.**  
   Чем выше максимальная высота снега, тем больше потенциальный объём талой воды весной.

2. **При весеннем потеплении этот запас быстро включается в сток.**  
   Особенно если одновременно есть сильные весенние осадки.

3. **Весенние осадки в Олёкминске — главный синхронный фактор:**  
   `precip_spring_z = +1.652`.  
   Это намного сильнее, чем снег `+0.497`.

4. **Суровая зима сформировала прочный лёд:**  
   `FDD_z = -0.910`.  
   Поэтому когда приходит большой объём воды от снега и дождей, русло ещё не готово свободно пропустить поток.

Итог:

> В Олёкминске высокий снежный покров в год ЧС работает как дополнительный водный резерв, который при весеннем увлажнении усиливает паводочную волну.

## Почему это не противоречит идее «снежного буфера»

Здесь важно разделить два разных показателя:

### 1. `max_snow_depth_z = +0.497`

Это **максимальная высота снежного покрова**.

Она отвечает на вопрос:

> Сколько снега накопилось к весне?

В синхронном анализе Олёкминска положительный сдвиг означает:

> перед/в годы всплеска ЧС снега было больше, и он мог дать больший объём талой воды.

---

### 2. `SO_z = +1.060`

Это уже **Surface Offset**, то есть показатель связи между температурой воздуха и температурой поверхности грунта.

Он отвечает на другой вопрос:

> Как снег, влажность и состояние поверхности изменяли теплообмен между воздухом и грунтом?

Вот именно к `SO_z = +1.060` относится мой вывод:

> «Такой сильный положительный сдвиг говорит о том, что перед годами с повышенным риском ЧС особенно важна роль поверхностного буфера: снега, влажности, состояния почвы и теплоизоляции».

То есть эту фразу корректнее относить **не к `max_snow_depth_z`**, а к **`SO_z`**.

## Главное отличие этих факторов

## Что НЕ является триггером всплеска

По твоим значениям не подтверждаются как главные триггеры:

```text
thaw_d_z = -1.045
ros_event_days_z = -0.457
ros_precip_sum_z = -0.672
ros_precip_max_z = -0.727
```

Почему?

Потому что все они имеют **отрицательный сдвиг**:

```text
Всплеск ЧС < Безопасно
```

То есть в годы катастрофических ЧС:

* оттепелей меньше;
* ROS-дней меньше;
* сумма ROS-осадков меньше;
* максимальный ROS-эпизод слабее.

Если бы они были основными триггерами всплеска, значения должны были бы быть положительными.

---

## Что может быть фоновым или вторичным фактором

ROS и оттепели могут сопровождать отдельные годы с одной ЧС.

Например, по `ros_event_days_z`:

```text
Всплеск: -0.600
Норма: +0.259
Безопасно: -0.143
```

То есть ROS-дни выше именно в группе **нормы**, где 1 ЧС, а не в группе всплесков.

Это означает:

> Rain-on-Snow может быть локальным триггером отдельных паводковых ситуаций, но не механизмом катастрофического всплеска.

То же касается `thaw_d_z`:

```text
Всплеск: -0.650
Норма: -0.128
Безопасно: +0.395
```

Здесь даже в норме значение выше, чем во всплеске, но безопасные годы имеют самый высокий уровень оттепельности. Значит, сами по себе оттепели не переводят систему в катастрофический режим.

## Физическая интерпретация для Якутска

Для Якутска это можно объяснить так:

1. **Мало оттепелей**  
   Ледовый покров не ослабляется зимой. Он сохраняет прочность до весны.

2. **Меньше снега**  
   Снежный покров хуже изолирует грунт и пойму от зимнего холода.

3. **Глубже промерзает поверхность**  
   Холод легче проникает в деятельный слой, пойменные участки и прибрежную зону.

4. **Формируется мерзлотный водоупор**  
   Весной талая вода хуже уходит в грунт.

5. **ROS отсутствует как сильный фактор**  
   Значит, катастрофа возникает не из-за кратковременного дождя по снегу, а из-за подготовленного ледово-мерзлотного состояния.

6. **Весенний приток воды становится опасным**  
   Когда приходит вода от таяния и/или осадков, она сталкивается с прочным льдом и промёрзшей поймой.

7. **Итог — заторы и подтопления**  
   Русло не успевает плавно очиститься, лёд разрушается резко.

# Общая интерпретация двух ALT-параметров

Оба показателя имеют отрицательный сдвиг:

```text
ALT_R42_z = -0.229
ALT_R43_z = -0.213
```

Это значит:

> Перед всплесками ЧС глубина сезонного протаивания на обеих площадках была ниже, чем перед безопасными годами.

На первый взгляд это может казаться странным: если лето теплее, почему протаивание не больше?

Но для мерзлоты это нормально. Глубина протаивания зависит не только от температуры воздуха, но и от:

* влажности грунта;
* теплоёмкости почвы;
* содержания льда;
* снежной изоляции;
* растительного и мохового покрова;
* способности грунта дренировать воду;
* зимнего промерзания.

Если грунт переувлажнён, он может прогреваться медленнее, потому что вода обладает высокой теплоёмкостью. В таком случае меньшее протаивание может быть не признаком «безопасности», а признаком **влажного, теплоёмкого, потенциально водоупорного состояния деятельного слоя**.

# Почему R42 и R43 нельзя трактовать одинаково

## R42

R42 — площадка в районе Якутск-Туймаада, где преобладают супеси и пески. Такие грунты обладают высокой водопроницаемостью: влага быстро дренирует и не задерживается в активном слое [1].

Поэтому слабый сдвиг:

```text
ALT_R42_z = -0.229
```

логичен. R42 не даёт сильного «эффекта памяти»: песчаный грунт быстрее пропускает воду, и отложенный риск на следующий год выражен слабо.

## R43

R43, наоборот, интереснее как мерзлотный индикатор. Здесь отрицательный `ALT_R43_z = -0.213` можно интерпретировать как признак того, что перед всплесками ЧС активный слой не просто глубже протаивает, а может находиться в более сложном переувлажнённом состоянии.

Особенно важно, что по Якутску рядом есть более сильные лаговые параметры:

```text
TDD_z = +0.878
MAGST_z = -0.458
TO_z = +0.606
N_f_z =
```

```text
N_f_z = +0.519
max_snow_depth_z = -0.563
```

Именно они лучше объясняют механизм, чем `ALT` сам по себе.

Получается такая цепочка:

```text
повышенное летнее тепло, TDD_z = +0.878
+ пониженная глубина протаивания на R43, ALT_R43_z = -0.213
+ отрицательный MAGST_z = -0.458
→ признак теплоёмкого / переувлажнённого состояния поверхности
+ положительный TO_z = +0.606
→ усиление вертикального температурного контраста
+ положительный N_f_z = +0.519
→ изменение зимнего промерзания
+ дефицит снежного покрова, max_snow_depth_z = -0.563
→ усиленное промерзание переувлажнённого слоя
→ криогенный водоупор
→ снижение инфильтрации весной N+1
→ рост поверхностного стока
→ заторы и подтопления
```

# Главное: почему отрицательный `ALT` не противоречит высокому `TDD`

Для Якутска в лаговом анализе:

```text
TDD_z = +0.878
ALT_R42_z = -0.229
ALT_R43_z = -0.213
```

На первый взгляд можно спросить:

> если лето теплее, почему глубина протаивания меньше?

Ответ: потому что **глубина протаивания — не только функция температуры воздуха**.

Она зависит от того, сколько энергии уходит:

1. на нагрев грунта;
2. на фазовые переходы воды и льда;
3. на испарение;
4. на прогрев влажного слоя;
5. на преодоление теплоёмкости переувлажнённого грунта.

Если грунт влажный или ледонасыщенный, значительная часть тепла расходуется не на увеличение глубины протаивания, а на фазовые процессы и прогрев воды. Поэтому при высоком `TDD_z` активный слой может не обязательно становиться глубже.

Для ВКР это можно объяснить так:

> Положительный сдвиг `TDD_z` указывает на повышенный летний тепловой запас, однако отрицательные сдвиги `ALT_R42_z` и `ALT_R43_z` показывают, что этот тепловой запас не реализовался в виде увеличения глубины протаивания. Вероятно, значительная часть энергии была затрачена на прогрев и фазовые переходы во влажном деятельном слое. Поэтому ALT в данном случае отражает не простую глубину деградации мерзлоты, а сложное водно-тепловое состояние грунта.

# Как интерпретировать роль `ALT_R43_z`

## Значения

```text
обычные ЧС → возможно, глубже протаивание
катастрофические ЧС → не глубже протаивание, а водно-тепловая перестройка слоя
```

> Отрицательный сдвиг `ALT_R43_z = -0.213` показывает, что перед катастрофическими всплесками ЧС глубина сезонного протаивания на R43 была ниже, чем перед безопасными годами. При этом в годы с одной ЧС значение `ALT_R43_z` было положительным. Следовательно, увеличение глубины протаивания может сопровождать локальные или единичные ЧС, но не объясняет переход к системному всплеску. Для катастрофических лет более значимым оказывается не абсолютная глубина протаивания, а состояние переувлажнённого и затем промёрзшего деятельного слоя.

## Тогда что означает такая комбинация?

```text
меньше оттепелей
+ меньше снега
+ меньше местных весенних осадков
+ слабый ROS
→ устойчиво холодный и малоснежный режим
→ глубокое промерзание руслово-пойменной зоны
→ прочный ледовый покров
→ позднее и резкое вскрытие нижней Лены
+ приход транзитной паводковой волны сверху
→ заторы и подтопления
```

То есть для Жиганска в синхронном анализе важна не локальная «добавка воды» через снег/осадки/ROS, а **ледово-русловая подготовка** и **транзитная волна**.

## Почему безопасные годы могут иметь больше ROS и осадков?

На первый взгляд странно:

```text
Безопасно:
ros_event_days_z = +0.788
ros_precip_max_z = +0.383
precip_spring_z = +0.255
```

А при всплесках:

```text
ros_event_days_z = -0.360
ros_precip_max_z = -0.470
precip_spring_z = -0.145
```

Например:

* лёд уже частично ослаблен;
* паводковая волна не совпала по времени;
* нет сильной транзитной волны сверху;
* местный сток не наложился на русловой затор;
* русло очистилось постепенно.

Тогда даже ROS или весенние осадки могут пройти без системной ЧС.

То есть для Жиганска важен не сам факт осадков или ROS, а состояние всей системы:

```text
ледовый покров
+ промерзание поймы
+ сроки вскрытия
+ транзитная волна
+ местные условия
```

# Важное уточнение: годовые осадки ≠ весенние осадки

Для Жиганска нельзя делать вывод, что «осадки вообще не важны».

Потому что в лаговом анализе у тебя есть сильный положительный показатель:

```text
precip_spring_z = +1.426
```

А здесь:

```text
precip_z = -0.401
```

Это означает:

> Важна не сумма осадков за год, а сезон их выпадения.

То есть если осадки приходятся на весну, они могут быть значимым фактором риска, а если смотреть годовую сумму целиком, сигнал размывается.

## Как это объяснить

В Нижней Лене весна и начало лета имеют особое значение:

* поверхность ещё холодная;
* мерзлота плохо принимает воду;
* испарение ограничено;
* вегетационный сезон короткий;
* вода может задерживаться в верхних горизонтах;
* при последующем промерзании формируется ледонасыщенный слой.

Поэтому:

```text
годовые осадки ↓
но весенние осадки ↑
→ общий год не влажный,
но критическое увлажнение произошло в нужный сезон
```

Это очень сильный вывод для ВКР.

## Почему снег отрицательный, а не положительный?

`max_snow_depth_z = -0.422` показывает, что перед всплесками ЧС снега меньше.

Для Жиганска это можно интерпретировать через теплоизоляцию:

```text
меньше снега
→ меньше теплоизоляции
→ холод глубже проникает в грунт и пойму
→ усиливается промерзание
→ формируется прочный ледово-мерзлотный барьер
→ весной вода хуже впитывается
```

То есть снег здесь работает не как «запас воды», а как **теплоизоляционный фактор**.

При большом снежном покрове грунт может быть теплее и менее глубоко промёрзшим.  
При малом снежном покрове грунт промерзает сильнее.

Поэтому в Жиганске отрицательный снег может быть связан с повышенным риском через механизм промерзания.

```text
thaw_d_z = +2.000
precip_spring_z = +1.426
MAGST_z = +0.400
TTOP_z = +0.530
precip_z = -0.401
max_snow_depth_z = -0.422
```

Вместе они дают не противоречивую, а очень интересную картину:

```text
зимние оттепели ↑
+ весенние осадки ↑
+ мерзлотный профиль теплее
но годовые осадки ↓
и снега меньше
```

То есть для Жиганска риск формируется не по схеме:

```text
много снега и много осадков весь год → ЧС
```

а по схеме:

```text
малоснежный / относительно сухой фон
+ отдельные критические сезонные нарушения
  - зимние оттепели
  - весеннее переувлажнение
  - тёплый мерзлотный профиль
→ нестабильный ледово-мерзлотный режим
→ заторы и подтопления
```

## Слайд. Жиганск: почему годовые осадки и снег не объясняют всплески

**Текст на слайде:**

* `precip_z = -0.401` — годовые осадки ниже перед всплесками ЧС.
* `max_snow_depth_z = -0.422` — снежный покров ниже перед всплесками.
* Следовательно, риск не связан с общим избытком влаги или снега.
* Важна сезонная структура:
  * `precip_spring_z = +1.426` — весеннее переувлажнение;
  * `thaw_d_z = +2.000` — зимние оттепели;
  * `MAGST_z = +0.400`, `TTOP_z = +0.530` — тёплый мерзлотный профиль.
* Вывод: Жиганск реагирует на отдельные сезонные нарушения, а не на годовую сумму осадков.

**Речь докладчика:**

«В лаговом анализе для Жиганска годовые осадки и снежный покров имеют отрицательные сдвиги. Это означает, что перед всплесками ЧС год в целом не был более влажным или более снежным. Поэтому простая схема “больше осадков и снега — выше риск” для Нижней Лены не работает.

Однако это не противоречит роли влаги. Наоборот, анализ показывает, что важна сезонная структура увлажнения. Весенние осадки имеют сильный положительный сдвиг, а зимние оттепели являются главным долгосрочным предиктором. На фоне более тёплого мерзлотного профиля это создаёт нестабильное ледово-мерзлотное состояние. Поэтому Жиганск реагирует не на годовую сумму осадков, а на критические сезонные нарушения — оттепели и весеннее переувлажнение.»

# Финальный вывод по Жиганску

> Для Жиганска отрицательные сдвиги `precip_z` и `max_snow_depth_z` показывают, что катастрофические всплески ЧС не формируются за счёт общего избытка годовых осадков или повышенного снежного запаса. Более значимым является сочетание малоснежного фона, тёплого мерзлотного профиля, зимних оттепелей и весеннего переувлажнения. Такая комбинация нарушает устойчивость нижнеленской ледово-мерзлотной системы и повышает вероятность заторов при прохождении паводковой волны сверху по течению.
